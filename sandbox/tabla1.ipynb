```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams["figure.figsize"] = (12,7)
sns.set_context("talk")
```

```python

file1 = open('tabla1.txt', 'r')
Lines = file1.readlines()

scores = []
for l in Lines:
    #print(l)
    scores.append(l.split(',')[2].split(' ')[3])
scores = [float(a) for a in scores]
```

```python
file2 = open('tabla2.txt', 'r')
Lines = file2.readlines()

scores2 = []
for l in Lines:
    #print(l)
    scores2.append(l.split(',')[2].split(' ')[3])
scores2 = [float(a) for a in scores2]
```

```python
file3 = open('tabla3.txt', 'r')
Lines = file3.readlines()

scores3 = []
for l in Lines:
    #print(l)
    scores3.append(l.split(',')[2].split(' ')[3])
scores3 = [float(a) for a in scores3]
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
def plot(scores, title = 'Title'):

    df = pd.DataFrame(scores, columns=["Reward"])
    temperature = df['Reward']
    t_average = temperature.rolling(window=50).mean()

    plt.plot(temperature, 'k-', label='Original', alpha=0.25)
    plt.plot(t_average, '#33cccc', label='Running average - 50 episodes')
    plt.ylabel('Sum of Rewards')
    plt.xlabel('Episode')
    plt.grid(linestyle=':')
    plt.fill_between(t_average.index, 0, t_average, color='#33cccc', alpha=0.05)
    plt.legend(loc='upper left')
    plt.title(title, fontsize=30 )
    plt.show()
```

```python
plot(scores, 'CNN Extra Reward')
```

```python
plot(scores2, 'CNN Model: Performance')
```

```python
plot(scores3, 'CNN + Sensors: Model Performance')
```
